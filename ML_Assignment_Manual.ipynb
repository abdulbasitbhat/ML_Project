```python
from sklearn.preprocessing import StandardScaler
import numpy as np
scaler = StandardScaler()
X = scaler.fit_transform(X)
X = np.insert(X,0,values=1,axis=1)
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1)

print(X_train)
print(Y_train)
```

```python
#Finding Beta's using Least square error fit for MLR
from numpy import linalg
A=X_train.T.dot(X_train)
B=np.linalg.inv(A)
C=B.dot(X_train.T)
beta=C.dot(Y_train)
print(beta)
```

```python
y_predict=X_test.dot(beta)
print(Y_test)
print(y_predict) 
```

```python
#Calculating Errors
error=Y_test-y_predict

square_error=np.power(error,2)
sum_square_error=np.sum(square_error)

#Mean Square error    sum((xi - xi')^2)/n
mean_square_error=sum_square_error/len(y_predict)
print(mean_square_error)

#Root mean Square error  (sum((xi - xi')^2)/n)^(0.5)
rms_error=np.sqrt(mean_square_error)
print(rms_error)

#Variance   sum((xi - xi')^2)/n
y_mean=np.mean(Y_test)
total_variance=np.sum((Y_test-y_mean)**2)
print(1-sum_square_error/total_variance)
```

```python
#Using Gradient Descent
lr = [0.0001,0.001,0.01,0.1,1,10]
lamdas = [10-15,10-10,10-5,10- 3,0,1,10,20]
Score = []
n=236
```

```python
#Updating Beta's using gradient descent, for various lamdas.
from numpy import linalg
from sklearn.metrics import r2_score

n = len(X_train)
for lamda in lamdas:
  for learning_rate in lr:
    beta=np.zeros(4)
    number_of_iterations=100
    for i in range(number_of_iterations):
      x0_gradient=0
      x1_gradient=0
      x2_gradient=0
      x3_gradient=0
      for j in range(len(X_train)):
        a=X_train[j,0]
        b=X_train[j,1]
        c=X_train[j,2]
        f=Y_train[j]
        x0_gradient+=(beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)-f)
        x1_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)-f)*a)
        x2_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)-f)*b)
        x3_gradient+=((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)-f)*c)
      beta[0]=beta[0]-learning_rate/n*x0_gradient
      beta[1]=beta[1]-learning_rate/n*x1_gradient
      beta[2]=beta[2]-learning_rate/n*x2_gradient
      beta[3]=beta[3]-learning_rate/n*x3_gradient
      y_pred = X_test.dot(beta)
      Score.append({'lamda':lamda,'lr':learning_rate,'score':r2_score(y_pred,Y_test),'beta':beta})
```

```python
#Finding Best Beta's that lead to least error (as given by max r2_score)
maxScore=-20
for i in Score:
  maxScore=max(maxScore,i['score'])
finalBeta = [final['beta'] for final in Score if final['score'] == maxScore]
finalLambda = [final['lamda'] for final in Score if final['score'] == maxScore]

finalBeta = finalBeta[0]
finalBeta
```

```python
Y_pred_grad = X_test.dot(finalBeta)
```

```python
error = np.asarray(Y_test).flatten() - Y_pred_grad

sum_square_error=np.sum(np.power(error,2))
mean_square_error=sum_square_error/len(Y_pred_grad)
rms_error=np.sqrt(mean_square_error)

print(rms_error)
```
